Importing Libraries for Data Correlation

```python

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'C:\Users\Ejben\Downloads\movies.csv')
```

Looking at the top results of the data.

```python
df.head()
```

Finding any missing data within the table

```python
# This loops through the dataframe and looks for any missing info

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
```

Looking at the data types for each of the columns.

```python
print(df.dtypes)
```

This code drops any rows with null values

```python
df = df.dropna(how='any',axis=0)
```

This command adds the release date column into a seperate column

```python
df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

This plot shows if any movies had a significantly higher gross compared to the others in this data set

```python
# Showing outliers

df.boxplot(column=['gross'])
```

Ordering the data by gross to compare it to the plot above
    The plot appears to be accurate with two movies making far more than the next two

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

This scatterplot shows the correlation between a movie's budget and it's gross.
    This shows that the higher the budget, the higher the gross.

```python
sns.regplot(x="gross", y="budget", data=df)
```

This scatterplot compares a movie's score to it's gross

```python
sns.regplot(x="score", y="gross", data=df)
```

These graphs show the correlation between all of the numeric values using three different methods

```python
df.corr(method ='pearson')
```

```python
df.corr(method ='kendall')
```

```python
df.corr(method ='spearman')
```

Using a matrix to show the correlation for all numeric values

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

Using factorize

```python
# this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

Showing the correlation values for name and runtime

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

Sorting the correlation values by their "correlation score".

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

Looking at the values with the highest correlation

```python
# high correlation is anything > 0.5

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

Assigns a numeric values to all string values to allow for more data correlations

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

Showing the correlation of all categories to each other

```python
df_numerized.corr(method='pearson')
```

Same as above but in a better visual manner

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
